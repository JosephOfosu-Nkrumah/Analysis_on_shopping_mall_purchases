```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default
```

```python
purchases = pd.read_excel("C:/Users/Joseph Ofosu-Nkrumah/Desktop/new/Data/purchase_data.xlsx")
```

```python
purchases
```

```python
#Total sales by month
total_sales_by_month = purchases.groupby("Month of Purchase")["Quantity Purchased"].sum() 

# Visualize total sales by month
plt.figure(figsize=(10, 6))
sns.barplot(x=total_sales_by_month.index, y=total_sales_by_month.values, palette="viridis")
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Quantity Purchased")
plt.xticks(rotation=45)
plt.savefig("Total Sales by Month.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Most purchased products
most_purchased_products = purchases.groupby("Product")["Quantity Purchased"].sum().nlargest(10)

#Visualize most purchased products
plt.figure(figsize=(10, 6))
sns.barplot(x=most_purchased_products.values, y=most_purchased_products.index, palette="magma")
plt.title("Top 10 Most Purchased Products")
plt.xlabel("Total Quantity Purchased")
plt.ylabel("Product")
plt.savefig("Top 10 Most Purchased Products.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Calculate total quantity purchased for each month
monthly_total_quantity = purchases.groupby("Month of Purchase")["Quantity Purchased"].sum()

# Select the top 4 months
top_4_months = monthly_total_quantity.nlargest(4)

# Visualize the top 4 months using a bar graph
plt.figure(figsize=(10, 6))
top_4_months.plot(kind='bar', color= ['skyblue','gold','violet','pink'])
plt.title("Top 4 Months with Highest Total Purchases")
plt.xlabel("Month")
plt.ylabel("Total Quantity Purchased")
plt.xticks(rotation=45)
plt.savefig("Top 4 Months with Highest Total Purchases.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Average price per product category
average_price_per_category = purchases.groupby("Product")["Price (per unit)"].mean().nlargest(10)

# Visualize average price per product category
plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_per_category.values, y=average_price_per_category.index, palette="plasma")
plt.title("Average Price per Product Category")
plt.xlabel("Average Price (per unit)")
plt.ylabel("Product Category")
plt.savefig("Average Price per Product Category.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Average quantity purchased per product category
average_quantity_per_category = purchases.groupby("Product")["Quantity Purchased"].mean().nlargest(10)

#Visualize average quantity purchased per product category
plt.figure(figsize=(10, 6))
sns.barplot(x=average_quantity_per_category.values, y=average_quantity_per_category.index, palette="rocket")
plt.title("Average Quantity Purchased per Product Category")
plt.xlabel("Average Quantity Purchased")
plt.ylabel("Product Category")
plt.savefig("Average Quantity Purchased per Product Category.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Distribution of total revenue by product category
total_revenue_per_category = purchases.groupby("Product")["Total Revenue"].sum().nlargest(10)

# Visualize distribution of total revenue by product category
plt.figure(figsize=(8, 8))
plt.pie(total_revenue_per_category, labels=total_revenue_per_category.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Total Revenue by Product Category")
plt.axis('equal')
plt.savefig("Distribution of Total Revenue by Product Category.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Distribution of total quantity purchased by product category
total_quantity_per_category = purchases.groupby("Product")["Quantity Purchased"].sum().nlargest(10)

# Visualize distribution of total quantity purchased by product category
plt.figure(figsize=(8, 8))
plt.pie(total_quantity_per_category, labels=total_quantity_per_category.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Total Quantity Purchased by Product Category")
plt.axis('equal')
plt.savefig("Distribution of Total Quantity Purchased by Product Category.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Monthly trend in total sales quantity
monthly_total_quantity = purchases.groupby("Month of Purchase")["Quantity Purchased"].sum()

#Visualize monthly trend in total sales quantity
plt.figure(figsize=(10, 6))
plt.plot(monthly_total_quantity.index, monthly_total_quantity.values, marker='o', color='b', linestyle='-')
plt.title("Monthly Trend in Total Sales Quantity")
plt.xlabel("Month")
plt.ylabel("Total Quantity Purchased")
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig("Monthly Trend in Total Sales Quantity.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```

```python
#Monthly trend in total revenue
monthly_total_revenue = purchases.groupby("Month of Purchase")["Total Revenue"].sum()

#Visualize monthly trend in total revenue
plt.figure(figsize=(10, 6))
plt.plot(monthly_total_revenue.index, monthly_total_revenue.values, marker='o', color='r', linestyle='-')
plt.title("Monthly Trend in Total Revenue")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig("Monthly Trend in Total Revenue.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```
